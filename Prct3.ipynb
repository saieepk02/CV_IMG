```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read images
img1 = cv2.imread("cvimg.jpg")
img2 = cv2.imread("cv2.jpg")

# Resize
img1 = cv2.resize(img1, (300, 300))
img2 = cv2.resize(img2, (300, 300))

# Convert to RGB
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# Operations
add_img = cv2.add(img1, img2)
sub_img = cv2.subtract(img1, img2)
blend = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)
mult = cv2.multiply(img1, img2)
div = cv2.divide(img1, img2)

# Display
plt.figure(figsize=(10,6))

plt.subplot(2,3,1), plt.imshow(img1), plt.title("Image 1")
plt.subplot(2,3,2), plt.imshow(img2), plt.title("Image 2")
plt.subplot(2,3,3), plt.imshow(add_img), plt.title("Add")
plt.subplot(2,3,4), plt.imshow(sub_img), plt.title("Subtract")
plt.subplot(2,3,5), plt.imshow(blend), plt.title("Blend")
plt.subplot(2,3,6), plt.imshow(mult), plt.title("Multiply")

plt.tight_layout()
plt.show()

# Division result
plt.imshow(div)
plt.title("Division")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read images
img1 = cv2.imread("cv2.jpg")
img2 = cv2.imread("cvimg.jpg")

# Resize
img1 = cv2.resize(img1, (300,300))
img2 = cv2.resize(img2, (300,300))

# Convert to RGB
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# Bitwise ops
or_img  = cv2.bitwise_or(img1, img2)
and_img = cv2.bitwise_and(img1, img2)
xor_img = cv2.bitwise_xor(img1, img2)
not_img = cv2.bitwise_not(img2)

# Show results
plt.subplot(2,3,1), plt.imshow(img1), plt.title("Img1")
plt.subplot(2,3,2), plt.imshow(img2), plt.title("Img2")
plt.subplot(2,3,3), plt.imshow(or_img),  plt.title("OR")
plt.subplot(2,3,4), plt.imshow(and_img), plt.title("AND")
plt.subplot(2,3,5), plt.imshow(xor_img), plt.title("XOR")
plt.subplot(2,3,6), plt.imshow(not_img), plt.title("NOT")

plt.show()
```
