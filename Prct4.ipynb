```python
import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread("cv2.jpg")

# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to Grayscale
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

# Histogram Equalization
img_eq = cv2.equalizeHist(img_gray)

# Display Images
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis("off")

plt.subplot(2, 3, 2)
plt.title("Grayscale Image")
plt.imshow(img_gray, cmap="gray")
plt.axis("off")

plt.subplot(2, 3, 3)
plt.title("Equalized Image")
plt.imshow(img_eq, cmap="gray")
plt.axis("off")

# Original Histogram
plt.subplot(2, 3, 4)
plt.title("Original Histogram")
plt.hist(img_gray.ravel(), bins=256, range=[0, 256])

# Equalized Histogram
plt.subplot(2, 3, 5)
plt.title("Equalized Histogram")
plt.hist(img_eq.ravel(), bins=256, range=[0, 256])

# Comparison Graph
plt.subplot(2, 3, 6)
plt.title("Comparison Graph")
plt.hist(img_gray.ravel(), bins=256, range=[0, 256], alpha=0.5, label="Original")
plt.hist(img_eq.ravel(), bins=256, range=[0, 256], alpha=0.5, label="Equalized")
plt.legend()

plt.tight_layout()
plt.show()
```
