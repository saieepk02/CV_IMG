```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 
# Read the image 
path = r"cv2.jpg" 
img = cv2.imread(path, 1) 
# Display original image 
plt.subplot(2, 2, 1) 
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) 
plt.title("Original Image") 
# Image Attributes 
print('Size of image:', img.size) 
print('Dimensions of image:', img.ndim) 
print('Datatype of image:', img.dtype) 
print('Shape of image:', img.shape) 
# Color Conversions 
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)   # BGR → RGB 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # BGR → Grayscale 
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)   # BGR → HSV 
plt.subplot(2, 2, 2) 
plt.imshow(rgb) 
plt.title("RGB Image") 
plt.subplot(2, 2, 3) 
plt.imshow(gray, cmap="gray") 
plt.title("Grayscale Image") 
plt.subplot(2, 2, 4) 
plt.imshow(hsv) 
plt.title("HSV Image") 
plt.tight_layout() 
plt.show() 
# Display Individual Color Channels 
r, g, b = cv2.split(rgb) 
plt.figure(figsize=(8, 4)) 
plt.subplot(1, 3, 1) 
plt.imshow(r, cmap='Reds') 
plt.title("Red Channel") 
plt.subplot(1, 3, 2) 
plt.imshow(g, cmap='Greens') 
plt.title("Green Channel") 
plt.subplot(1, 3, 3) 
plt.imshow(b, cmap='Blues') 
plt.title("Blue Channel") 
plt.tight_layout() 
plt.show()
```

```python
#B – Basic Image Processing Operations (Resize, Rotate, Flip, Crop): 
#RESIZE AN IMG 
resize_img = cv2.resize(img, (0, 0), fx=2, fy=1) 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
resize_img_rgb = cv2.cvtColor(resize_img, cv2.COLOR_BGR2RGB) 
h1, w1 = img.shape[:2] 
h2, w2 = resize_img.shape[:2] 
plt.figure(figsize=(10, 5)) 
# Original image 
plt.subplot(1, 2, 1) 
plt.imshow(img_rgb) 
plt.title(f"Original Image - Dimensions: {w1}x{h1}") 
# Resized image 
plt.subplot(1, 2, 2) 
plt.imshow(resize_img_rgb) 
plt.title(f"Resized Image - Dimensions: {w2}x{h2}") 
plt.show() 
#ROTATING AN IMG 
plt.figure(figsize=(10, 5)) 
#cv2.ROTATE_90_CLOCKWISE 
rot_img1 = cv2.rotate(img_rgb, cv2.ROTATE_90_CLOCKWISE) 
plt.subplot(1,3,1) 
plt.imshow(rot_img1) 
plt.title("ROTATE_90_CLOCKWISE") 
#cv2.ROTATE_180 
rot_img = cv2.rotate(img_rgb, cv2.ROTATE_180) 
plt.subplot(1,3,2) 
plt.imshow(rot_img) 
plt.title("ROTATE_180") 
#cv2.ROTATE_90_COUNTERCLOCKWISE 
rot_img2 = cv2.rotate(img_rgb,cv2.ROTATE_90_COUNTERCLOCKWISE) 
plt.subplot(1,3,3) 
plt.imshow(rot_img2) 
plt.title("ROTATE_90_COUNTERCLOCKWISE") 
plt.show() 
# Flips 
flip_h = cv2.flip(img_rgb, 1)   # Horizontal 
flip_both = cv2.flip(img_rgb, -1) # Both (180°) 
flip_v = cv2.flip(img_rgb, 0)  # Vertical 
# Subplots with separate axes 
fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharex=False, sharey=False) 
axs[0].imshow(flip_h) 
axs[0].set_title("Horizontal Flip") 
axs[1].imshow(flip_both) 
axs[1].set_title("Flip Both (180°)") 
axs[2].imshow(flip_v) 
axs[2].set_title("Vertical Flip") 
plt.show() 
#CROPING AN IMG 
crop_img=img_rgb[50:130,100:300] 
plt.imshow(crop_img) 
plt.show()
```
